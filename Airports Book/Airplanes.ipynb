```python
#import dependencies and csv files
import pandas as pd 

airports = pd.read_csv('airports', names=["index", "airport_name", "city", "country", "airport", "idk_id", "lat", "lng",
                                         "altitude", "timezone", "DST", "Tz", "airport_type", "source"])
delays = pd.read_csv('469260606_122019_5347_airline_delay_causes.csv')
```

```python
airports.head(2)
```

```python
delays.head(2)
```

```python
#clean & drop un-needed columns

airports.dropna()
airports1 = airports[['airport_name','city', 'country','airport', 'lat','lng']]
airports1.head()
```

```python
#clean & drop un-needed columns
delays.dropna()
delays.head()
```

```python
aviation = pd.merge(delays, airports1, on='airport', how='outer')
aviation.head()
```

```python
aviation.to_csv('aviation_data.csv')
```
